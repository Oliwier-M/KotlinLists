```python
import pandas as pd
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
df = pd.read_csv("data.csv", skiprows=4)
```

```python
df.head()
```

```python
df.replace(-99, np.NAN, inplace=True)
```

```python
df
```

```python
df.interpolate(inplace=True)
```

```python
df
```

```python
df['Date'] = pd.to_datetime(df['Date'], format='%Y%m')
```

```python
df.set_index(['Date'], inplace=True)
df.head()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(df.index, df['Value'], label='Temperature Time Series', color='blue')
plt.title('Average Temperature Time Series')
plt.xlabel('Date')
plt.ylabel('Temperature')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.hist(df['Value'], bins=20, color='green')
plt.title('Histogram of Temperature')
plt.xlabel('Temperature')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.kdeplot(df['Value'], color='orange', shade=True)
plt.title('Kernel Density Plot of Temperature')
plt.xlabel('Temperature')
plt.ylabel('Density')
plt.grid(True)
plt.show()
```
